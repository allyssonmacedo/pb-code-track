## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
df1 = pd.read_csv('Datasets/gasolina_2010+.csv')
df1
```

```python
df2 = pd.read_csv('Datasets/gasolina_2000+.csv')
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

### 2.Investigue as colunas e entenda o conjunto de dados usando o head() e info()

```python
df3.info()
```

### 3. Selecione a terceira entrada da coluna DATA INICIAL e verifique seu tipo.

```python
df3['DATA INICIAL'][2]
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df3['DATA INICIAL'] = pd.to_datetime(df3['DATA INICIAL'])
df3['DATA FINAL'] = pd.to_datetime(df3['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
df3['anomes'] = df3['DATA FINAL'].apply(lambda x: f"{x.year}") + df3['DATA FINAL'].apply(lambda x: f'/{x.month:02d}')
df3['anomes']
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
df3.value_counts('PRODUTO')
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_gasolina = df3[df3['PRODUTO'] == 'GASOLINA COMUM']
df_gasolina
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
df_gasolina[df_gasolina['anomes'] == '2008/08']['PREÇO MÉDIO REVENDA'].mean()
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
df_gasolina[(df_gasolina['anomes'] == '2014/05') & (df_gasolina['ESTADO'] == 'SAO PAULO')]['PREÇO MÉDIO REVENDA'].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
df_gasolina[df_gasolina['PREÇO MÉDIO REVENDA'] > 5][['ESTADO', 'anomes', 'PREÇO MÉDIO REVENDA']]
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_2012 = df_gasolina[df_gasolina['DATA FINAL'].apply(lambda x: x.year) == 2012]


df_2012[df_2012['REGIÃO'] == 'SUL']['PREÇO MÉDIO REVENDA'].mean()
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df_gasolina['ano'] = df_gasolina['DATA FINAL'].apply(lambda x: x.year)
df_gasolina['mes'] = df_gasolina['DATA FINAL'].apply(lambda x: x.month)

df_rio = df_gasolina[df_gasolina['ESTADO'] == 'RIO DE JANEIRO']

df_rio_mes = df_rio.groupby('anomes')[['PREÇO MÉDIO REVENDA', 'mes']].last()
```

```python
(df_rio_mes[df_rio_mes['mes'] == 12] / df_rio_mes[df_rio_mes['mes'] == 12].shift(1) - 1) * 100
```

```python
((df_rio_mes[df_rio_mes['mes'] == 12] / df_rio_mes[df_rio_mes['mes'] == 12].shift(1) - 1) * 100).rolling(3).sum()
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_max = df_gasolina.groupby('anomes').max()['PREÇO MÉDIO REVENDA']
df_min = df_gasolina.groupby('anomes').min()['PREÇO MÉDIO REVENDA']
```

```python
df_max
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff['abs_diff'] = df_max - df_min
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
```

```python
idx_max = df_gasolina.groupby('anomes')['PREÇO MÉDIO REVENDA'].idxmax()
idx_min = df_gasolina.groupby('anomes')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
idx_max['2013/06']
```

```python
df_diff['max'] = df_max
df_diff['min'] = df_max
```

```python
df_diff
```

```python
df_diff['ESTADO_MAX'] = df_gasolina.loc[idx_max, :][['ESTADO', 'anomes']].groupby('anomes').last()
df_diff['ESTADO_MIN'] = df_gasolina.loc[idx_min, :][['ESTADO', 'anomes']].groupby('anomes').last()
```
