```python
import pandas as pd
import openpyxl
```

```python
tb_equipamentos = pd.read_excel('controle_equipamentos.xlsx', header = 2)
tb_veiculos = pd.read_excel('veiculos_unidades.xlsx', header = 3)

print(tb_equipamentos.columns)
print(tb_veiculos.columns)

print(tb_equipamentos.dtypes)
print(tb_veiculos.dtypes)


tb_equipamentos['FROTA'] = pd.to_numeric(tb_equipamentos['FROTA'])
tb_veiculos['EQUIPAMENTO'] = pd.to_numeric(tb_veiculos['EQUIPAMENTO'])

display(tb_veiculos.head())
display(tb_equipamentos.head())
```

```python
tabela = pd.merge(tb_veiculos, tb_equipamentos, left_on = 'EQUIPAMENTO', right_on = 'FROTA', how = 'outer')
display(tabela.head())

tabela.to_excel('EquipamentosUnid_outer.xlsx', index = False)
```
